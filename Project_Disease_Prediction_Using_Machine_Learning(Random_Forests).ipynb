#### Train|Test Split

```python
X = df.drop('prognosis', axis=1)
y = df['prognosis']
```

```python
y.head()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

#### Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
help(RandomForestClassifier)
```

```python
model = RandomForestClassifier(n_estimators = 3, random_state=101)
```

```python
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
```

```python
preds
```

```python
preds[55]
```

##### Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix, accuracy_score        
```

```python
confusion_matrix(y_test,preds)
```

```python
accuracy_score(y_test, preds)
```

##### Feature Importance

```python
model.feature_importances_
```

##### Choosing More Number of Trees

```python
test_error = []

for n in range(1,5):
    model  = RandomForestClassifier(n_estimators=n)
    model.fit(X_train, y_train)
    test_preds = model.predict(X_test)
    test_error.append(1- accuracy_score(y_test, test_preds))
    
```

```python
plt.plot(range(1,5), test_error )
```

We can see from the above graph that at n_estimators >=2, we get 100 percent accuracy.
